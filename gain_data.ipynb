```python
code = pd.read_csv("data/code_list.csv",usecols=['ts_code',"list_date"])
```

```python
count = 0
for i in code["ts_code"].tolist()[count:2]:
    pro = ts.pro_api()
    df = pro.weekly(ts_code = i,start_date = "19950501",end_date = "20190501",fields = "trade_date,close,pct_chg")
    df.to_csv("trade_data/"+i+".csv")
    count += 1
    print(i,end="")
    print(count)
```
