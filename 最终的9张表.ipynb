# 形成期为1周，持有期为1周

## （1）在此之前，先将收益率数据合成一张表

```python
code_list = pd.read_csv("cleaned_fin_data/cleaned_code_list.csv",usecols=["ts_code"])
union = pd.read_csv("trade_data/000001.SZ.csv",index_col=0,usecols=["trade_date","pct_chg"])
union.columns = ["000001.SZ"]
for i in code_list.ts_code[1:]:
    pct_chg = pd.read_csv("trade_data/"+i+".csv",index_col=0,usecols=["trade_date","pct_chg"])
    pct_chg.columns = [i]
    union = pd.concat([union,pct_chg],axis=1).fillna(0.0)
    print(i)
```

```python
union.to_csv("trade_data/union.csv")
```

## （2）union加一列season

```python
union.insert(0,"season",0)
```

```python
for i in range(len(union)):
    temp = str(union.trade_date[i])
    if int(temp[4:6])<3 or int(temp[4:6])==3:
        temp = str(int(temp[:4])-1)+"1231"
    elif int(temp[4:6])<6 or int(temp[4:6])==6:
        temp = temp[:4]+"0331"
    elif int(temp[4:6])<9 or int(temp[4:6])==9:
        temp = temp[:4]+"0630"
    else:
        temp = temp[:4]+"0930"
    print(i)
    union.season[i]=temp
```

```python
union = pd.read_csv("trade_data/union.csv")
```

```python
#生成表，并得到列名
day_map = total_map.get(str(union.season[125]))
column = list(day_map.keys())
j_1_return = pd.DataFrame(columns = column)

k_period = list(range(len(union)))
k_period = k_period[72:][::2]
for i in k_period:
    j_rate_ls = []
    k_rate_ls = []
    day_map = total_map.get(str(union.season[i]))
    
    for j in range(25):
        key = "p"+str(j+1)
        port = day_map.get(key)
        j_rate,k_rate = 0,0
        for p in port:
            j_rate += union[p][i-1]
            k_rate += union[p][i]
        j_rate /= len(port)
        k_rate /= len(port)
        j_rate_ls.append(j_rate)
        k_rate_ls.append(k_rate)
    j_1_return.loc[len(j_1_return)]=j_rate_ls
    j_1_return.loc[len(k_1_return)]=k_rate_ls
```

## 建立循环

```python
union = pd.read_csv("trade_data/union.csv")
#生成表，并得到列名
day_map = total_map.get(str(union.season[125]))
column = list(day_map.keys())
d = [1,2,3,4,6,8,12]
for n in d:
    print(n)
    j_d_return = pd.DataFrame(columns = column)

    k_period = list(range(len(union)))
    k_period = k_period[72:][::n+1]
    for i in k_period:
        j_rate_ls = []
        k_rate_ls = []
        day_map = total_map.get(str(union.season[i]))

        for j in range(25):
            key = "p"+str(j+1)
            port = day_map.get(key)
            j_rate,k_rate = 0,0
            for p in port:
                j_rate += (sum(union[p][i-n:i])/n)
                k_rate += union[p][i]
            j_rate /= len(port)
            k_rate /= len(port)
            j_rate_ls.append(j_rate)
            k_rate_ls.append(k_rate)
        j_d_return.loc[len(j_d_return)]=j_rate_ls
        j_d_return.loc[len(j_d_return)]=k_rate_ls
    j_d_return.to_csv("style_momentrum_effect/j_"+str(n)+"_return.csv")
    print("style_momentrum_effect/j_"+str(n)+"_return.csv")
    print(np.mean(j_d_return.p1))
```
