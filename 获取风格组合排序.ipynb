### 市盈率数据

```python
date = []
for i in range(94):
    temp = ""
    temp += str(2018-(i-1)//4)
    month = str((4-(i-1)%4)*3)
    if month == "6" :
        day = "30"
    elif  month == "9":
        day ="30"
    else:
        day = "31"
    temp += "{:0>2}".format(month)
    temp += day
    date.append(temp)

mother = pd.DataFrame({"date":date})

code_list = pd.read_csv("cleaned_fin_data/cleaned_code_list.csv",usecols=["ts_code","list_date"])
for c in code_list["ts_code"]:
    bps = pd.read_csv("cleaned_fin_data/"+c+".csv",usecols=["date","bps"])
    mother[c] = bps.bps
    print(c,end=" ")
    print(code_list[code_list.ts_code == c].index.tolist())

ls = [len(mother)-1,len(mother)-2,len(mother)-3]    
mother.drop(ls,axis = 0,inplace=True)
mother.to_csv("dictionary/pe.csv")
```

## 先根据资产表中每一行数据，划分5个组合

## 建立循环

```python
ls = list(range(3442))
ls.pop(0)
asset = pd.read_csv("dictionary/assets.csv",usecols=ls)
```

```python
first_map = {}
for i in range(len(asset)):
    seq = asset.loc[i,:]
    seq.drop("date",inplace=True)
    seq.drop(seq[seq == 0.0].index,inplace=True)
    seq.sort_values(ascending=False,inplace=True)
    num = len(seq)-1-seq.tolist().count(0)
    each = num//5
    last = each + (num % 5)
    dic = {}
    for j in range(5):
        ls = []
        start = each*j
        end = each*(j+1) if j !=4 else each*j+last
        ls.append(seq.index.tolist()[start:end])
        dic[str(j+1)] = ls
    first_map[asset.date[i]] = dic
    print(i)
```

## 第二步：根据市盈率表格划分出25个组合

```python
ls = list(range(3442))
ls.pop(0)
pe = pd.read_csv("dictionary/pe.csv",usecols=ls)
```

```python
total_map = {}
sec_map = {}
for  p in range(len(pe)):
    day = first_map.get(asset.date[p])

    for j in range(5):
        ls = []
        asset_order = day.get(str(j+1))[0]
        for i in asset_order:
            ls.append(pe[i][0])
        group = pd.DataFrame({"name":asset_order,"pe":ls})
        group.sort_values(by=["pe"],ascending=False,inplace=True)

        each = len(group)//5
        last = each + len(group)%5
        for k in range(5):
            if k  == 4:
                sec_map["p"+str(j*5+5)]=group.name.tolist()[each*4:]
            else:
                num = "p"+str(j*5+k+1)
                sec_map[num]=group.name.tolist()[each*k:each*(k+1)]
    total_map[str(asset.date[p])] = sec_map
    print(asset.date[p],end=" ")
    print(p)

```
